```python
import pandas as pd

data = {
"Customer_ID":["C101","C102","C103","C104","C105","C106","C107","C108","C109","C110"],
"Age":[22,35,28,45,32,26,50,29,41,23],
"City":["Bhubaneswar","Delhi","Mumbai","Kolkata","Chennai","Bhubaneswar","Delhi","Mumbai","Chennai","Kolkata"],
"Transaction_Amount":[4500,12000,30000,2000,50000,7000,150000,2500,80000,1000],
"Transaction_Type":["Debit","Credit","Debit","Debit","Credit","Debit","Debit","Debit","Credit","Debit"],
"Account_Balance":[35000,88000,150000,12000,300000,40000,900000,22000,600000,8000]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,5))
sns.histplot(df["Transaction_Amount"], bins=5, kde=True)
plt.title("Transaction Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="City", y="Transaction_Amount", data=df)
plt.title("City Wise Transaction Amount")
plt.show()
```
